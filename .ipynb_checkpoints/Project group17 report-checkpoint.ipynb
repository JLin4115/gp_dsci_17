## 2.5 Predictive Data Analysis

In order to make our analysis to be reproducible, we use `set.seed` at the beginning of the analysis. 

Then, we create the `k_vals` data frame with the neighbors variable containing values from 1 to 25 (stepping by 2) using the `seq` function. 

Also, we change the subscribe from logic to factor by using `mutate()`. Use the `initial_split` function to split the data into a training (75%) and test set (25%). Set the `strata` argument to the class label variable which is subscribe. The `training` and `testing` functions then extract the training and testing data sets into two separate data frames. 

```r
# Setting seed = 1111 to ensure reproducible results. DO NOT CHANGE
set.seed(1111)

k_vals <- tibble(neighbors = seq(from = 1, to = 25, by = 2))

players_knn <- players_clean |>
  mutate(subscribe = as_factor(subscribe)) 

# Making Initial Split
players_split <- initial_split(players_knn, prop = 0.75, strata = subscribe)
players_training <- training(players_split)
players_testing  <- testing(players_split)

# Training data using cross validation
players_recipe <- recipe(subscribe ~ experience + played_hours, data = players_training) |>
  step_scale(all_predictors()) |>
  step_center(all_predictors())

# Create a nearest_neighbors model specification, with neighbors = tune(), storing in knn_spec
knn_spec <- nearest_neighbor(weight_func = "rectangular",neighbors = tune()) |>
  set_engine("kknn") |>
  set_mode("classification")

# Create a 5-fold cross-validation object
players_vfold <- vfold_cv(players_training, v = 5, strata = subscribe)


knn_results <- workflow() |>
  add_recipe(players_recipe) |>
  add_model(knn_spec) |>
  tune_grid(resamples = players_vfold, grid = k_vals) |>
  collect_metrics()

# Finding best K
accuracies <- knn_results |>
  filter(.metric == "accuracy")

best_k <- accuracies |>
  arrange(desc(mean)) |>
  slice(1) |>
  pull(neighbors)

# Predicting on test data
knnspec_final <- nearest_neighbor(weight_func = "rectangular", neighbors = best_k) |>
  set_engine("kknn") |>
  set_mode("classification")

knn_fit <- workflow() |>
  add_recipe(players_recipe) |>
  add_model(knnspec_final) |>
  fit(data = players_training)

players_test_predictions <- predict(knn_fit, players_testing) |>
  bind_cols(players_testing)

# Evaluating tuned model accuracy
players_test_accuracy <- players_test_predictions |>
  metrics(truth = subscribe, estimate = .pred_class) |>
  filter(.metric == "accuracy")

players_test_accuracy
```

```r
# Visualization of best K
options(repr.plot.width = 8, repr.plot.height = 7)
best_k_plot <- accuracies |> 
ggplot(aes(x = neighbors, y = mean)) +
    geom_point() + 
    geom_line() + 
    scale_x_continuous(breaks = seq(0, 25, by = 1)) +  
    scale_y_continuous(limits = c(0.4, 1.0)) +
    labs(x = "Number of Neighbors (K)", y = "Accuracy Estimate", title = "Figure 4: Number of Neighbours (K) vs Accuracy Estimate") +
    theme(text = element_text(size=15))
best_k_plot
```

Figure 4 shows that the accuracy estimate of the classifier increases with the number of neighbors until 70% ~ accuracy at 19 neighbors. There is then a minor decrease in accuracy when increasing the number of neighbors to 20, before the curve flattens out.

# 3. Discussion

## 3.1 Summary of Findings

Our research question for this project was if experience level and play time could act as predictors for subscription status to a game-related newsletters in the players.csv dataset. From our exploratory and predictive analysis, it is observed that these variables are overall weak predictors of player subscription status. 

To begin, when performing exploratory analysis, figure 1 showed that there was little distinction in subscription levels for different experience classes of players. There was no real trend, as all of the subscription percentiles within each class was ~75%. This was the first indication that experience level likely did not have significant influence on the subscription status of a player. This would also contribute to a less accurate predictor because it has less of a pattern to go off of.

However, when looking at playtime, there was a clear distinction in subscription levels for players who played less versus more; figure 2 highlighted that players who had an average playtime of 0.5 hours during data collection were not subscribed to game-related newsletters. Contrastingly, it showed that players who had an average playtime of 7+ hours during the data collection were subscribed. This visualization had a clear pattern, which indicated that playtime would be a better predictor for the knn model.

Figure 3 showed that certain patterns were present for some experience classes, while no trends were observed for other classes. For example for classes 1 and 4, players that had a higher playtime appeared to have higher subscription levels. However, for classes 2, 3, and 5, despite some outlier data points, playtime did not appear to be correlated with subscription status. This confirmed that although playtime was a better predictor than experience level, there was no clear trend in the subscription status based on both predictors combined. This indicated that the predictor that we would create was likely to have low accuracy. 

Finally, our predictive analysis confirmed that our chosen variables failed to serve as predictors for subcription status. After finding the number of neighbors that resulted in the highest prediction accuracy, the accuracy was still only 73%. For a prediction model, this is extremely poor. The predictor will be incorrect about its predictions over a quater of the time! 

## 3.2 Alignment with expectations

Our group hypothesized that both experience and playtime during the data collection would be strong predictors for subscription status. Thus, a knn model for the `players.csv` data would have relatively high accuracy. 

Specifically, we expected higher experience classes to be correlated with higher subscription rates, while lower experience classes would be correlated with lower subscription rates. This is because more experienced players are more likely to be highly engaged in the games that they play, and thus want to be informed on game-related events through newsletter subscriptions. However, from figure one, it is observed that subscription levels were relatively consistent across all experience classes. In fact, the most experienced players actually had the lowest percentage of subscription, which was surprising. 

This feeds into one of the major uncertainties of the research project during data collection and analysis; experience level is subjective. Survey takers can all have different perceptions of the different classes, and there was no standardization in place to remove perception bias in this regard. Someone may have considered themself a veteran because they had an extensive “Minecraft phase” years prior, but do not play as much anymore. This may have also been why there didn’t seem to be a large correlation between playtime and experience level in figure 3. Beginners actually had a higher play time on average, which may be due to the fact that they are becoming more interested in the game, and thus increasing their engagement with the game and the study itself. 

This uncertainty may explain the results seen in figure 1 that contrasted the expectations we had for an experience level versus subscription status plot. 

What did align was our expectations about playtime being a strong predictor of subscription status: We hypothesized that a higher average playtime would result in a higher likelihood for subscription for a game-related newsletter, while lower average playtime would result in the opposite. This was seen to be a trend in figure 2, supporting our hypothesis. 

Although playtime was observed to be quite strong as a predictor (as expected, the uncertainty associated with the experience level variable introduced some predictive complications into our model, reducing its ability to predict subscription status. This contributed to our knn predictor being quite ineffective, which was also counter to our expectations. 

## 3.3 Impact of Findings 

The findings from this analysis have several important implications for the marketing companies/teams and corporations associated with the data set. First, the results show that experience level offers very limited value as a segmentation variable. Since subscription rates remained relatively consistent across all experience categories, marketers cannot rely on experience level to predict which players are likely to subscribe to newsletters. This means that campaigns designed to target players based on their experience level alone are unlikely to be effective or generate a meaningful return on investment.

While playtime showed a more noticeable relationship with subscription behaviour, particularly among players with very low or very high playtime, it still did not produce strong predictive accuracy when used in combination with experience level. This suggests that playtime may offer some contextual insight, such as differentiating highly engaged players from disengaged ones, but it is not a reliable standalone predictor for targeted marketing. It would be worth combining with another strongly predictive variable to get accurate predictions of the types of consumers that should be targeted for subscriptions. 

## 3.4 Implications for Future Research 

The experience level variable was observed to be problematic for this type of data analysis. It is an inherently subjective variable that can be skewed by the perception of survey-takers, which reduces its effectiveness as a predictor in our model. In future research to reduce this, experience level data could be collected in a more quantitative manner that reduced bias. The research project actually had in-game data collection that looked at mouse movement and clicking patterns. This type of data collection could similarity be used for the experience level variable, without the player knowing; in other words, a model could be created to analyze things like in-game reaction time, speed of mouse movement and clicks, etc, and create an experience level predictor based on these variables. This would remove a large amount of bias in the experience level variable, making it a more viable predictor for subscription status prediction. 

Another potential extension of this research project at its root would be to collect additional data regarding behavioural variables that are more directly tied to marketing engagement, such as players’ interactions with promotional material and login frequency.
